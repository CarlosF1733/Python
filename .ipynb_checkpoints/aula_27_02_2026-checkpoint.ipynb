```python
login = "mary_anne"
secret = "****"

while True:
    user_login = input("\nLogin: ")
    if user_login != login:
        print("Fail on login")
        continue
    password = input("\nPass: ")
    if password != secret:
        print("Fail on pass")
        continue
    break
print("Access Granted")
```

```python
#handling exceptions
def x_dividing_by_y(x,y):
    return x/y
print(x_dividing_by_y(10,2))
```

```python
#handling exceptions
def x_dividing_by_y(x,y):
    try:
        return x/y

    except TypeError:
        print("You need to add a valid number")

    except ZeroDivisionError:
        print(f"You cannot divide {x} by {y}.")  
        
print(x_dividing_by_y(5,0))
```
